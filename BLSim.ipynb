```python
goals_table = []

for i in teams:
    # Get all games from team i
    # Home games
    home = res_1920[res_1920.iloc[:, 0] == i]
    away = res_1920[res_1920.iloc[:, 1] == i]
    home_goals = home.iloc[:,2].sum()/17
    home_against = home.iloc[:,3].sum()/17
    away_goals = away.iloc[:,3].sum()/17
    away_against = away.iloc[:,2].sum()/17
    goals = (home_goals + away_goals)/2
    against = (home_against + away_against)/2
    dict_row = {'Team': i, 'goals': goals, 'against': against, 'home_goals': home_goals, 'home_against': home_against,
                'away_goals': away_goals, 'away_against': away_against}
    goals_table.append(dict_row)
    
goals_table = pd.DataFrame(goals_table)
goals_table
```

```python
goals_mean = goals_table["goals"].mean()
home_goals_mean = goals_table['home_goals'].mean()
home_against_mean = goals_table['home_against'].mean()
print(goals_mean, home_goals_mean, home_against_mean)
```

```python
files = os.listdir("Data/1BL")

goals_table_complete = []

for file in files:
    df = pd.read_csv("Data\\1BL\\" + file, encoding = 'unicode_escape')
    results = df.iloc[:,2:6]
    teams = pd.unique(results['HomeTeam'])
    season = file[2:6]
    #print(season)
    for team in teams:

        # Get all games from team i
        # Home games
        home = results[results.iloc[:, 0] == team]
        # Away games
        away = results[results.iloc[:, 1] == team]
        
        home_goals = home.iloc[:,2].sum()/17
        home_against = home.iloc[:,3].sum()/17
        away_goals = away.iloc[:,3].sum()/17
        away_against = away.iloc[:,2].sum()/17
        goals = (home_goals + away_goals)/2
        against = (home_against + away_against)/2
        dict_row = {'Season': season, 'Team': team, 'goals': goals, 'against': against, 'home_goals': home_goals,
                    'home_against': home_against, 'away_goals': away_goals, 'away_against': away_against}
        goals_table_complete.append(dict_row)
    
goals_table_all = pd.DataFrame(goals_table_complete)
goals_table_all
```

```python
data_rows = []
seasons = goals_table_all["Season"].unique()
#season = seasons[20]

for season in seasons:
    goals_mean = goals_table_all[goals_table_all["Season"] == season]["goals"].mean()
    against_mean = goals_table_all[goals_table_all["Season"] == season]["against"].mean()
    home_goals_mean = goals_table_all[goals_table_all["Season"] == season]["home_goals"].mean()
    home_against_mean = goals_table_all[goals_table_all["Season"] == season]["home_against"].mean()
    away_goals_mean = goals_table_all[goals_table_all["Season"] == season]["away_goals"].mean()
    away_against_mean = goals_table_all[goals_table_all["Season"] == season]["away_against"].mean()
    row = {"season": season, "goals": goals_mean, "against": against_mean, "home_goals": home_goals_mean,
           "home_against": home_against_mean, "away_goals": away_goals_mean, "away_against": away_against_mean,
          "goals_against_diff": (home_goals_mean-home_against_mean)}

    data_rows.append(row)

goal_means_table = pd.DataFrame(data_rows)

#for season in seasons:
print(goal_means_table["goals_against_diff"].mean())
goal_means_table
```

```python
goals_mean = goal_means_table["goals"]
goals_home = goal_means_table["home_goals"]
goals_against = goal_means_table["home_against"]
x = np.arange(2000,2021)
plt.plot(x, goals_mean, label = "Total")
plt.plot(x, goals_home, label = "Heim")
plt.plot(x, goals_against, label = "Auswaerts")
plt.legend()
plt.grid(True)
```
